```python
#Country count
print("Number of countries recorded in the dataset:", data['Country'].nunique())
data=data.round(2)
```

```python
#Correcting the column names
data=data.rename(columns={'Density\n(P/Km2)':'Density (P/Km2)'})
data['Country'] = data['Country'].str.replace("S�����������", "Sao Tome and Principe")
```

```python
#Removing the '%', ',' symbol
data['Agricultural Land( %)'] = data['Agricultural Land( %)'].str.replace('%', '')
data['CPI Change (%)']= data['CPI Change (%)'].str.replace('%','')
data['Forested Area (%)']= data['Forested Area (%)'].str.replace('%','')
data['Gross primary education enrollment (%)']= data['Gross primary education enrollment (%)'].str.replace('%','')
data['Gross tertiary education enrollment (%)']= data['Gross tertiary education enrollment (%)'].str.replace('%','')
data['Population: Labor force participation (%)']= data['Population: Labor force participation (%)'].str.replace('%','')
data['Tax revenue (%)']= data['Tax revenue (%)'].str.replace('%','')
data['Total tax rate']= data['Total tax rate'].str.replace('%','')
data['Unemployment rate']= data['Unemployment rate'].str.replace('%','')
data['Out of pocket health expenditure']= data['Out of pocket health expenditure'].str.replace('%','')
data['Land Area(Km2)']=data['Land Area(Km2)'].str.replace(',','')
data['Armed Forces size']=data['Armed Forces size'].str.replace(',','')
data['Co2-Emissions']=data['Co2-Emissions'].str.replace(',','')
data['Population']=data['Population'].str.replace(',','')
data['Urban_population']=data['Urban_population'].str.replace(',','')
data['Minimum wage']=data['Minimum wage'].str.replace('$','')
data['GDP']=data['GDP'].str.replace('$','')
data['GDP']=data['GDP'].str.replace(',','')
data['Gasoline Price']=data['Gasoline Price'].str.replace('$','')
data['CPI']=data['CPI'].str.replace(',','').replace('.','')
data['Density (P/Km2)']=data['Density (P/Km2)'].str.replace(',','')
```

```python
#Converting the data type to float data type and numeric
data['CPI']=data['CPI'].astype('float')
data['Density (P/Km2)']=pd.to_numeric(data['Density (P/Km2)'])
data['Agricultural Land( %)'] = data['Agricultural Land( %)'].astype('float')
data['CPI Change (%)']= data['CPI Change (%)'].astype('float')
data['Forested Area (%)']= data['Forested Area (%)'].astype('float')
data['Gross primary education enrollment (%)']= data['Gross primary education enrollment (%)'].astype('float')
data['Gross tertiary education enrollment (%)']= data['Gross tertiary education enrollment (%)'].astype('float')
data['Population: Labor force participation (%)']= data['Population: Labor force participation (%)'].astype('float')
data['Tax revenue (%)']= data['Tax revenue (%)'].astype('float')
data['Total tax rate']= data['Total tax rate'].astype('float')
data['Unemployment rate']= data['Unemployment rate'].astype('float')
data['Out of pocket health expenditure']= data['Out of pocket health expenditure'].astype('float')
data['Land Area(Km2)']=data['Land Area(Km2)'].astype('float')
data['Armed Forces size']=pd.to_numeric(data['Armed Forces size'])
data['Co2-Emissions']=pd.to_numeric(data['Co2-Emissions'])
data['Population']=data['Population'].astype('float')
data['Urban_population']=pd.to_numeric(data['Urban_population'])
data['Minimum wage']=data['Minimum wage'].astype('float')
data['Gasoline Price']=data['Gasoline Price'].astype('float')
data['GDP']=pd.to_numeric(data['GDP'])

```

```python
import plotly.express as px
fig = px.choropleth(data_frame=data,
                    locations='Country',  
                    locationmode='country names',  
                    color='GDP', 
                    title='GDP by Country',
                    hover_name='Country',  
                    color_continuous_scale=px.colors.sequential.Plasma,)
fig.update_layout(autosize=False, width=800, height=700)
fig.show()
plt.show()
```
